```python
from gapminder import gapminder
import pandas as pd
import seaborn as sns
```

```python
gapminder.head()
```

```python
data = gapminder[gapminder['country'] == 'Afghanistan']
```

```python
data.plot(title="Life Expectancy in Afghanistan",
         y = 'pop',
         x = 'year')
```

### Area Map

```python
data.plot(title="Life Expectancy in Afghanistan",
         y = 'pop',
         x = 'year',
         kind = 'area')
```

### Scatter Plot

```python
data.plot(title="Life Expectancy in Afghanistan",
         y = 'pop',
         x = 'year',
         kind = 'scatter')
```

## Using Histogram to Compare Across Countries

```python
data_2007 = gapminder[gapminder['year'] == 2007]
```

```python
data_2007.plot(title='Populations in 2007',
              kind = 'hist',
              y='pop',
              bins=100)
```

### Using Scatterplot to View Trends Between Variables

```python
gapminder.plot(title = 'GDP per Capita vs. Life Expectancy',
         kind = 'scatter',
         x = 'gdpPercap',
         y = 'lifeExp')
```

### Logging Variables for Exploratory Data Analysis

```python
gapminder.plot(title = 'GDP per Capita vs. Life Expectancy',
         kind = 'scatter',
         x = 'gdpPercap',
         y = 'lifeExp',
         logx = True)
```

### Using Color to Distinguish Variables

```python
gapminder.continent.unique()
```

```python
## create dictionary of color keys

cont_color = {'Asia':'red',
             'Africa':'orange',
             'Americas':'navy',
             'Europe':'green',
             'Oceania':'teal'}

## create new column in gapminder dataset
gapminder['color'] = [cont_color[country] for country in gapminder['continent']]
```

```python
gapminder.plot(title = 'GDP per Capita vs. Life Expectancy',
         kind = 'scatter',
         x = 'gdpPercap',
         y = 'lifeExp',
         logx = True,
         c = gapminder['color'])
```

```python
g = sns.PairGrid(gapminder, diag_sharey = False)

g.map_lower(sns.kdeplot)

g.map_upper(sns.scatterplot)

g.map_diag(sns.kdeplot, lw=3)
```
